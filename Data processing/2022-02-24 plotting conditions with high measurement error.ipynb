```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('./paper.mplstyle.py')
```

```python
for sheet_ in [0.79, 0.623, 0.606, 0.535, 0.423, 0.405, 0.335]:
    df = pd.read_excel('./2022-02-24 replicate data for conditions with high measurement error.xlsx', sheet_name = str(sheet_))
    t = list(df.iloc[:, 0])[1:-1]
    r1 = list(df.iloc[:, 5])[1:-1]
    r2 = list(df.iloc[:, 6])[1:-1]
    r3 = list(df.iloc[:, 7])[1:-1]
    mean = list(df.iloc[:, 8])[1:-1]
    
    fig = plt.figure(figsize = (3.5,3))
    
    for replicate in [r1, r2, r3]:
        plt.plot(t, replicate, marker = 'None', linestyle = 'solid', color = 'dimgrey')
        
    plt.plot(t, mean, marker = 'None', linestyle = 'solid', color = 'steelblue')
    plt.xlabel('Time (min)')
    plt.ylabel('Normalized readout')
    plt.savefig('./replicate time course ' + str(sheet_) + '.svg')
    

    
```
